# Ex5.27

```python
from vector_drawing import *

dino_vectors = [(6,4), (3,1), (1,2), (-1,5), (-2,5), (-3,4), (-4,4),
    (-5,3), (-5,2), (-2,2), (-5,1), (-4,0), (-2,1), (-1,0), (0,-3),
    (-1,-4), (1,-4), (2,-3), (1,-2), (3,-1), (5,1)
]

draw(
    Points(*dino_vectors),
    Polygon(*dino_vectors)
)
```

```python
from draw3d import draw3d, Points3D, Segment3D

def polygon_segments_3d(points,color='blue'):
    count = len(points)
    return [Segment3D(points[i], points[(i+1) % count],color=color) for i in range(0,count)]

dino_3d = [(x,y,1) for x,y in dino_vectors]

draw3d(
    Points3D(*dino_3d, color='blue'),
    *polygon_segments_3d(dino_3d)
)
```

```python
# x-2y-2
from transforms import multiply_matrix_vector

parallel_move_matrix = ((1,0,-2),
                        (0,1,-2),
                        (0,0,1))
                        
moved_dino_3d = [multiply_matrix_vector(parallel_move_matrix, vector) for vector in dino_3d]
draw3d(
    Points3D(*moved_dino_3d, color='blue'),
    *polygon_segments_3d(moved_dino_3d)
)
```

```python
moved_dino_2d = [(x,y) for (x,y,z) in moved_dino_3d]
draw(
    Points(*dino_vectors,color='C0'),
    Polygon(*dino_vectors,color='C0'),
    Points(*moved_dino_2d, color='C3'),
    Polygon(*moved_dino_2d, color='C3'),
)
```

```python
from math import pi, cos, sin

# ...
transform_matrix = ((cos(pi/4),0,2),
                     (0,sin(pi/4),2),
                     (0,0,1))
                        
transformed_dino_3d = [multiply_matrix_vector(transform_matrix, vector) for vector in dino_3d]
draw3d(
    Points3D(*transformed_dino_3d, color='blue'),
    *polygon_segments_3d(transformed_dino_3d)
)
```

```python
transformed_2d = [(x,y) for (x,y,z) in transformed_dino_3d]

draw(
    Points(*dino_vectors,color='C0'),
    Polygon(*dino_vectors,color='C0'),
    Points(*transformed_2d, color='C3'),
    Polygon(*transformed_2d, color='C3'),
)
```
