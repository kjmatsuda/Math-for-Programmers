# Plane Vector Arithmetic

```python
def add(v1,v2):
    return (v1[0] + v2[0], v1[1] + v2[1])
```

```python
dino_vectors2 = [add((-1.5,-2.5), v) for v in dino_vectors]
```

```python
draw(
    Points(*dino_vectors, color=blue),
    Polygon(*dino_vectors, color=blue),
    Points(*dino_vectors2, color=red),
    Polygon(*dino_vectors2, color=red)
)
```

```python
arrows = [Segment(tip,tail,color=black) for (tip,tail) in
         zip(dino_vectors2, dino_vectors)]
draw(
    Points(*dino_vectors, color=blue),
    Polygon(*dino_vectors, color=blue),
    Points(*dino_vectors2, color=red),
    Polygon(*dino_vectors2, color=red),
    *arrows
)
```

```python
def my_distance(v1, v2):
    return length(my_subtract(v1, v2))

print(my_distance((5,4), (4,6)))
```

```python
def my_perimeter(vectors):
    sum = 0
    print(len(vectors))
    for i in range(len(vectors)):
        if i >= len(vectors)-1:
            sum += my_distance(vectors[i], vectors[0])
        else:
            sum += my_distance(vectors[i], vectors[i+1])
    return sum

my_perimeter([(1,0),(1,1),(0,1),(0,0)])
```

## 2.4 ベクトルの集合を座標変換する

ex2.42

```python

def my_rotate(angle, vectors):
    polar_coords = [to_polar(v) for v in vectors]
    rotated_polar_coords = [(l, theta + angle) for (l, theta) in polar_coords]
    return [to_cartesian(polar_v) for polar_v in rotated_polar_coords]

my_rotate(pi, [(1, 1)])
```

```python
rotated_dino_vectors = my_rotate(pi / 4, dino_vectors)

draw(
    Points(*dino_vectors, color=blue),
    Polygon(*dino_vectors, color=blue),
    Points(*rotated_dino_vectors, color=red),
    Polygon(*rotated_dino_vectors, color=red)
)
```

ex2.43

```python
def my_regular_polygon(n):
    polar_vectors = []
    angle_step = 2 * pi / n
    for i in range(n):
        polar_vectors.append((1, angle_step * i))
    return [to_cartesian(v) for v in polar_vectors]
    
my_regular_polygon(3)
```

```python
n_polygon_vectors = my_regular_polygon(7)
draw(
    Points(*n_polygon_vectors, color=blue),
    Polygon(*n_polygon_vectors, color=blue)
)
```
