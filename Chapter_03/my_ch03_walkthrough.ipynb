# Ex3.2

```python
draw3d(
      # 上の面
       Segment3D((1, 1, 1),(-1,1,1))
      ,Segment3D((-1, 1, 1),(-1,-1,1))
      ,Segment3D((-1, -1, 1),(1,-1,1))
      ,Segment3D((1, -1, 1),(1,1,1))
    
      # 下の面
      ,Segment3D((1, 1, -1),(-1,1,-1))
      ,Segment3D((-1, 1, -1),(-1,-1,-1))
      ,Segment3D((-1, -1, -1),(1,-1,-1))
      ,Segment3D((1, -1, -1),(1,1,-1))
    
      # 上下の面をつなぐ線分
      ,Segment3D((1, 1, 1),(1,1,-1))
      ,Segment3D((-1, 1, 1),(-1,1,-1))
      ,Segment3D((-1, -1, 1),(-1,-1,-1))
      ,Segment3D((1, -1, 1),(1,-1,-1))
     )
```

## 本に記載の回答

```python
pm1 = [1, -1]

edges = [((-1, y, z), (1, y, z)) for y in pm1 for z in pm1] + \
        [((x, -1, z), (x, 1, z)) for x in pm1 for z in pm1] + \
        [((x, y, -1), (x, y, 1)) for x in pm1 for y in pm1]

draw3d(*[Segment3D(*edge) for edge in edges])
```

```python
from math import sin, cos, pi
vs = [(sin(pi * t / 6), cos(pi * t / 6), 1.0 / 3) for t in range(0, 24)]

sum_vector = add(vs[0], vs[1])

for i in range(2, 24):
    sum_vector = add(sum_vector, vs[i])

print(sum_vector)

# 終点
draw3d(
       Arrow3D(vs[0], color = red)
       ,Arrow3D(sum_vector, color = purple)
       )
```

## 本に記載の回答

```python
from math import sin, cos, pi
vs = [(sin(pi * t / 6), cos(pi * t / 6), 1.0 / 3) for t in range(0, 24)]

running_sum = (0, 0, 0)
arrows = []

for v in vs:
    next_sum = add(running_sum, v)
    arrows.append(Arrow3D(next_sum, running_sum))
    running_sum = next_sum
print(running_sum)
draw3d(*arrows)
```

# Ex3.9

```python
# 本に記載の回答
def vectors_with_whole_number_length(max_coord=100):
    for x in range(1, max_coord):
        for y in range(1, x + 1):
            for z in range(1, y + 1):
                if length((x, y, z)).is_integer():
                    yield (x, y, z)

for vector in vectors_with_whole_number_length():
    print(vector)
```
